# Analyzing Train and Test Datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Extracting Data

```python
modelnet_test_file = '../logs/data_analysis/modelnet_test.csv'
modelnet_train_file = '../logs/data_analysis/modelnet_train.csv'
shapenet_file = '../logs/data_analysis/shapenet_full_train.csv'

labels = ["modelnet train (lk)", 
          "modelnet test (lk)", 
          "shapenet (c3po)"]
```

```python
modelnet_train_df = pd.read_csv(modelnet_train_file)
modelnet_test_df = pd.read_csv(modelnet_test_file)
shapenet_df = pd.read_csv(shapenet_file)

modelnet_train = modelnet_train_df.to_dict('list')
modelnet_test = modelnet_test_df.to_dict('list')
shapenet = shapenet_df.to_dict('list')
```

## Induced Rotation Errors

```python
rerr_modelnet_train = dict()
rerr_modelnet_test = dict()
rerr_shapenet = dict()

rerr_modelnet_train["modelnet train (lk)"] = modelnet_train["rerr"]
rerr_modelnet_test["modelnet test (lk)"] = modelnet_test["rerr"]
rerr_shapenet["shapenet (c3po)"] = shapenet["rerr"]

rerr_modelnet_train_df = pd.DataFrame.from_dict(rerr_modelnet_train)
rerr_modelnet_test_df = pd.DataFrame.from_dict(rerr_modelnet_test)
rerr_shapenet_df = pd.DataFrame.from_dict(rerr_shapenet)
```

```python
conca = pd.concat([rerr_modelnet_train_df.assign(dataset="modelnet train (lk)"),
                  rerr_modelnet_test_df.assign(dataset="modelnet test (lk)"),
                  rerr_shapenet_df.assign(dataset="shapenet (c3po)")])
```

```python
sns.set(style="darkgrid")
sns.kdeplot(conca, bw_adjust=0.1, cumulative=True, common_norm=False)
plt.xlabel('Axis-angle (rad)')
```

## Induced Translation Errors

```python
terr_modelnet_train = dict()
terr_modelnet_test = dict()
terr_shapenet = dict()

terr_modelnet_train["modelnet train (lk)"] = modelnet_train["terr"]
terr_modelnet_test["modelnet test (lk)"] = modelnet_test["terr"]
terr_shapenet["shapenet (c3po)"] = shapenet["terr"]

terr_modelnet_train_df = pd.DataFrame.from_dict(terr_modelnet_train)
terr_modelnet_test_df = pd.DataFrame.from_dict(terr_modelnet_test)
terr_shapenet_df = pd.DataFrame.from_dict(terr_shapenet)
```

```python
conca = pd.concat([terr_modelnet_train_df.assign(dataset="modelnet train (lk)"),
                  terr_modelnet_test_df.assign(dataset="modelnet test (lk)"),
                  terr_shapenet_df.assign(dataset="shapenet (c3po)")])
```

```python
sns.set(style="darkgrid")
sns.kdeplot(conca, bw_adjust=0.1, cumulative=True, common_norm=False)
plt.xlabel('Translation error')
```
