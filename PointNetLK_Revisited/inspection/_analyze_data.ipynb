# Analyzing Train and Test Datasets

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Extracting Data

```python
labels = ["modelnet train (lk)", 
          "modelnet test (lk)", 
          "shapenet.sim.easy", "shapenet.sim.medium", "shapenet.sim.hard",
          "shapenet.real.easy", "shapenet.real.medium", "shapenet.real.hard"]
```

```python
files = []

# modelnet train
files.append('../data_analysis/modelnet_train.csv')

# modelnet test
files.append('../data_analysis/modelnet_test.csv')

# shapenet.sim.easy 
files.append('../data_analysis/shapenet.sim.easy_test.csv')

# shapenet.sim.medium 
files.append('../data_analysis/shapenet.sim.medium_test.csv')

# shapenet.sim.hard
files.append('../data_analysis/shapenet.sim.hard_test.csv')

# shapenet.real.easy 
files.append('../data_analysis/shapenet.real.easy_test.csv')

# shapenet.real.medium 
files.append('../data_analysis/shapenet.real.medium_test.csv')

# shapenet.real.hard
files.append('../data_analysis/shapenet.real.hard_test.csv')
```

```python
data = dict()

for idx, file in enumerate(files):
    label = labels[idx]
    df = pd.read_csv(file)
    data[label] = df.to_dict('list')
```

## Induced Rotation Errors

```python
rerrs = []

for label in labels: 
    dict_ = dict()
    dict_[label] = data[label]['rerr']
    df = pd.DataFrame.from_dict(dict_)
    rerrs.append(df.assign(dataset=label))

conca = pd.concat(rerrs)
```

```python
sns.set(style="darkgrid")
sns.kdeplot(conca, bw_adjust=0.1, cumulative=True, common_norm=False)
plt.xlabel('Axis-angle (rad)')
```

## Induced Translation Errors

```python
terrs = []

for label in labels: 
    dict_ = dict()
    dict_[label] = data[label]['terr']
    df = pd.DataFrame.from_dict(dict_)
    terrs.append(df.assign(dataset=label))

conca = pd.concat(terrs)
```

```python
sns.set(style="darkgrid")
sns.kdeplot(conca, bw_adjust=0.1, cumulative=True, common_norm=False)
plt.xlabel('Axis-angle (rad)')
```
